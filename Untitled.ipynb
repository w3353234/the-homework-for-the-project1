```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("new_data.csv",encoding='utf-8')
test = pd.read_csv("test_set.csv",encoding='utf-8')
```

```python
train = pd.DataFrame(train)
test = pd.DataFrame(test)
```

```python
target=['winner']
unrelated=['gameId','gameDuration','seasonId']
```

```python
train_features=train.drop(target,axis=1)
train_features=train_features.drop(unrelated,axis=1)
train_labels=train[target]

test_features=test.drop(target,axis=1)
test_features=test_features.drop(unrelated,axis=1)
test_labels=test[target]
```

# 1.Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(criterion='entropy',splitter='best')
```

```python
dt.fit(train_features,train_labels)
```

```python
dt_pred = dt.predict(test_features)
```

```python
test_labels=test_labels.values
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test_labels,dt_pred)
```

# SVC

```python
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
svc = make_pipeline(StandardScaler(), SVC(gamma='auto',probability=True))
svc.fit(train_features, train_labels)
```

```python
svc_pred=svc.predict(test_features)
```

```python
svc.score(test_features,test_labels)
```

```python
from sklearn.metrics import f1_score
f1_score(test_labels,svc_pred)
```

# Ensemble of  Model

```python
dt_pred
```

```python
svc_pred
```

```python
_1 = 0
_2 = 0
final_pred = []
#初始化票数和最终结果

for i in range(0,len(dt_pred)):
    #遍历每个值
    #由于三个分类器精度都差不多，所以直接投票得到结果
    if dt_pred[i]==1:
        _1+=1
    if dt_pred[i]==2:
        _2+=1
            
    if svc_pred[i]==1:
        _1+=1
    if svc_pred[i]==2:
        _2+=1
    if _1>=_2:
        final_pred.append(1)
    else:
        final_pred.append(2)
            
    _1=0
    _2=0
```

```python
final_pred
```

```python
error = test_labels - final_pred
ensemble_score = np.power(np.sum(np.power(error,2)),1/2)/len(error)
```
